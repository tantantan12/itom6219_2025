# Data Import

```python
df=pd.read_csv("serp.csv")
df
```

TFIDF

```python
docs=df['title']
#Convert a collection of raw documents to a matrix of TF-IDF features.
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(docs)


tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_df
```

Topic Modeling

```python

# Apply NMF

nmf_model = NMF(n_components=7, random_state=0)
nmf_model.fit(tfidf)
W = nmf_model.fit_transform(tfidf)  # Document-topic matrix

# Display topics
feature_names = tfidf_vectorizer.get_feature_names_out()
topic_names=[]
for topic_index in range(len(nmf_model.components_)):
    topic = nmf_model.components_[topic_index]
    # Get the indices of the top 3 words (largest values in the topic)
    sorted_indices = topic.argsort()  # sorts from smallest to largest
    top_indices = sorted_indices[-4:]  # get the last 3 (top 3 words)
    # Reverse to make it largest to smallest
    top_indices = top_indices[::-1]
    # Get the actual word names for these indices
    top_words = []
    for i in top_indices:
        top_words.append(feature_names[i])
    # Join the top words into a single string
    top_words_string = " ".join(top_words)
    # Print and save
    print("Topic #{}:".format(topic_index))
    print(top_words_string)
    topic_names.append(top_words_string)

topic_df = pd.DataFrame(W, columns=topic_names)
topic_df
```

## Linear Regression (Basic)

```python
import pingouin as pg

# Combine X and y into a single dataframe
df_model = topic_df.copy()
df_model['rank'] = df['rank']

# Run linear regression
result = pg.linear_regression(df_model.drop(columns='rank'), df_model['rank'])

# Round coef and pval to 3 decimal places
result[['names', 'coef', 'pval']] = result[['names', 'coef', 'pval']].round(3)

# Display the rounded result
result[['names', 'coef', 'pval']]

result
```

# How I got the data?

```python
API_KEY =  
KEYWORDS = ["AI marketing", "AI Promotion", "AI Copy Generator"]
RESULTS_PER_PAGE = 100  # Max per page allowed by SerpAPI
MAX_RESULTS = 200

all_results = []

for keyword in KEYWORDS:
    total_collected = 0
    start = 0

    while total_collected < MAX_RESULTS:
        params = {
            "engine": "google",
            "q": keyword,
            "api_key": API_KEY,
            "hl": "en",
            "gl": "us",
            "num": RESULTS_PER_PAGE,
            "start": start
        }

        response = requests.get("https://serpapi.com/search", params=params)
        data = response.json()

        if "organic_results" not in data:
            print(f"No more results for keyword: {keyword}")
            break

        search_metadata = data.get("search_metadata", {})
        search_parameters = data.get("search_parameters", {})
        search_info = data.get("search_information", {})

        total_results = search_info.get("total_results")
        latency = search_metadata.get("total_time_taken")
        location = search_parameters.get("location", "us")
        language = search_parameters.get("hl", "en")

        for result in data["organic_results"]:
            title = result.get("title")
            snippet = result.get("snippet")
            rank = result.get("position")
            link = result.get("link")
            domain = urlparse(link).netloc if link else None

            all_results.append({
                "keyword": keyword,
                "rank": rank,
                "title": title,
                "snippet": snippet,
                "url": link,
                "domain": domain,
                "total_results": total_results,
                "latency": latency,
                "location": location,
                "language": language
            })

        total_collected += len(data["organic_results"])
        start += RESULTS_PER_PAGE
        time.sleep(2)  # Avoid hitting rate limits

# Create DataFrame
df = pd.DataFrame(all_results)

# Display or save
print(df)
# Domain Popularity
df['domain_popularity'] = df['domain'].map(df['domain'].value_counts())
```
